```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

df1 = pd.read_csv('client_cpu_data.csv')
df2 = pd.read_csv('client_gpu_data.csv')
df3 = pd.read_csv('rd_cpu_data.csv')
df4 = pd.read_csv('rd_gpu_data.csv')
```

```python

def boxplot(df,date,columns):
    #change type to datetime object 
    df[date] = pd.to_datetime(df[date])
    #create a column based on month
    df['month'] = df[date].dt.strftime('%b-%Y')
    #plot
    for i in columns:
        sns.boxplot(x='month', y=i, data=df)
        plt.show()
```

```python
list1 = ['queued_time', 'running_time', 'cpu_num', 'cpu_hour']
boxplot(df1,'date',list1)
```

```python
list2 = ['queued_time', 'running_time', 'gpu_num', 'gpu_hour']
boxplot(df2,'date',list2)
```

```python
list3 = ['queued_time', 'running_time', 'cpu_num', 'cpu_hour']
boxplot(df3,'date',list3)
```

```python
list4 = ['queued_time', 'running_time', 'gpu_num', 'gpu_hour']
boxplot(df4,'date',list4)
```

```python
def scatterplot(df,date,columns):
    #change type to datetime object 
    df[date] = pd.to_datetime(df[date])
    #create a column based on month
    df['month'] = df[date].dt.strftime('%b-%Y')
    #plot
    for i in columns:        
        sns.set(style='whitegrid')
        sns.scatterplot(x="month", y=i, data=df)
    plt.show()
```

```python
scatterplot(df1,'date',list1)
```
